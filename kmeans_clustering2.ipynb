Segmentation of customers using kmeans clustering algorithm

```python
#Importing libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
%matplotlib inline
```

```python
#Reading data file
customer = pd.read_csv("Cust_Segmentation.csv")
customer.head()
```

```python
#We can see that "Address column" is a categorical variable so we can drop it
cust = customer.drop('Address', axis=1)
cust.head()
```

```python
#Normalizing the data

from sklearn.preprocessing import StandardScaler
X = cust.values[:,1:]
#nan_to_num() function is used when we want to replace nan(Not A Number) with zero and inf with finite numbers in an array. It returns (positive) infinity with a very large number and negative infinity with a very small (or negative) number.
X = np.nan_to_num(X)
cdata = StandardScaler().fit_transform(X)
cdata
```

```python
#Modeling
cnum = 3
k_means = KMeans(init ="k-means++", n_clusters = cnum, n_init = 10)
k_means.fit(X)
labels = k_means.labels_
print(labels)
```

```python
cust["Cluster labels"] = labels
cust.head()
```

```python
#checking the centroid values by averaging the features in each cluster.
cust.groupby('Cluster labels').mean()
```

```python
#Distribution of customers based on age and income

area = np.pi * (X[:,1])**2
plt.scatter(X[:,0], X[:,3],s=area,c =labels.astype(np.float),alpha=0.5)
plt.xlabel('Age of the customer')
plt.ylabel('Monthly Income')
plt.show()
```

```python
#Visualisation

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(1, figsize=(8,6))
plt.clf()
ax = Axes3D(fig, rect= [0,0,.95,1], elev = 48, azim =134)

ax.set_xlabel('Education')
ax.set_ylabel('Age')
ax.set_zlabel('Income')
ax.scatter(X[:, 1], X[:, 0], X[:, 3], c= labels.astype(np.float))
```

k-means will partition your customers into 3 clusters. The customers in each cluster are similar to each other demographically. 
3 customer groups:
--> AFFLUENT, EDUCATED AND OLD AGED
--> MIDDLE AGED AND MIDDLE INCOME
--> YOUNG AND LOW INCOME
